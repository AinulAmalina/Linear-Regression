```python
#Linear Regression model for intercept=false

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import sklearn.linear_model
import sklearn.metrics
import sklearn.preprocessing
import sklearn.pipeline
import sklearn.datasets

sns.set()
```

```python
rng = np.random.RandomState(1)
X = 100 * rng.rand(50)
y = 26 * X - 722 + rng.randn(50)*200
```

```python
X.shape
```

```python
y.shape
```

```python
#Choose model hyperparameters

from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
model
```

```python
#Arrange data into a features matrix and target vector
print("Shape of x: {}".format(X.shape))
print(X)
```

```python
print("Shape of y: {}".format(y.shape))
print(y)
```

```python
x = X[:,np.newaxis]
print("Shape of x: {}".format(X.shape))
print(X)
```

```python
#Fit the model to data
model.fit(x, y)
```

```python
# intercept, i.e. beta_0
model.intercept_
```

```python
# slope, i.e. beta_1
model.coef_
```

```python
#Visually inspect the fit
yfit = model.predict(x)
```

```python
print("Shape of yfit: {}".format(yfit.shape))
print(yfit)
```

```python
model.intercept_ + model.coef_ * x[0]
```

```python
#plot the raw data and the predicted values using matplotlib
plt.scatter(x, y);
plt.plot(x, yfit, color='red');
```

```python
#The slope and intercept of the data 

print("Model slope:    ", model.coef_)
print("Model intercept:", model.intercept_)
```

```python
#compute the mean squared error with scikit-learn
from sklearn.metrics import mean_squared_error
mean_squared_error(y_true=y, y_pred=yfit)
```

```python

#R2 score
from sklearn.metrics import r2_score
r2_score(y_true=y, y_pred=yfit)
```
